```python
# Create a temp data frame to show which state has the most accidents
total_accidents = df.groupby("State") \
    .size() \
    .reset_index(name="Accident_Count") \
    .sort_values("Accident_Count", ascending=False)

total_accidents
```

```python
# Create a graph about accidents while grouping by city
accidents_by_cities = df.groupby("City") \
    .size() \
    .reset_index(name="Accident_Count") \
    .sort_values("Accident_Count", ascending=False)

accidents_by_cities.head(20).plot(
    kind="bar",
    x="City",
    y="Accident_Count",
    figsize=(12,4)
)
```

```python
# Create a graph about accidents while grouping by weather
weather_condition = df.groupby("Weather_Condition") \
    .size() \
    .reset_index(name="Accident_Count") \
    .sort_values("Accident_Count", ascending=False)
    
weather_condition.head(20).plot(
    kind="bar",
    x="Weather_Condition",
    y="Accident_Count", 
    figsize=(12,4)
)
```
