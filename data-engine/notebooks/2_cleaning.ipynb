```python
import pandas as pd

df = pd.read_csv("../datasets/raw/US_Accidents_March23.csv", nrows=100000)
df.head()
```

```python
# Create a copy 
copy_df = df
copy_df
```

```python
# Eliminate blank spaces at the bigginning and at the end 
copy_df = copy_df.apply(
    lambda col: col.str.strip() 
    if col.dtype == "object" 
        else col
)
```

```python
print(copy_df.columns)
```

```python
# Eliminate unnecessary columns form the copy
copy_df = copy_df.drop(["End_Lat", "End_Lng"], axis=1)
```

```python
copy_df = copy_df.drop(
    [
        "ID", "Source", "Description", 
        "Zipcode", "Country", "Timezone",
        "Country", "Timezone", "Airport_Code",
        "Weather_Timestamp", "End_Time"
     ],
    axis=1
)
```

```python
copy_df = copy_df.drop(
    [
        'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight',
        'Wind_Chill(F)', 'Pressure(in)', 'Wind_Direction',
        'County', 'Distance(mi)'
    ],
    axis=1
)
```

```python
# Verify if the dropped columns doesn't exist at copy
print(copy_df.columns)
```

```python
# convert the date string into date date
copy_df["Start_Time"] = pd.to_datetime(copy_df["Start_Time"], errors="coerce")

# Ahora sí puedes verificar
print("Tipo de dato actual:", copy_df['Start_Time'].dtype)
```

```python
copy_df['Hour'] =  copy_df['Start_Time'].dt.hour
copy_df['Month'] = copy_df['Start_Time'].dt.month
copy_df['Weekday'] =  copy_df['Start_Time'].dt.weekday
```

```python
bool_cols = [
    'Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 
    'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 
    'Traffic_Signal', 'Turning_Loop'
]
existing_bools = [col for col in bool_cols if col in copy_df.columns]
copy_df[existing_bools] = copy_df[existing_bools].astype(int)
```

```python
copy_df['Sunrise_Sunset'] = copy_df['Sunrise_Sunset'].fillna('Day')
copy_df['Is_Day'] = copy_df['Sunrise_Sunset'].apply(lambda x: 1 if x == 'Day' else 0)

copy_df = copy_df.drop('Sunrise_Sunset', axis=1)
```

```python
print(copy_df.head())
```

```python
copy_df = copy_df.drop("Start_Time", axis=1)
print(copy_df.columns)
```

```python
print("Verificando columnas antes de guardar:")
print(copy_df.columns)
```

```python
path = "../datasets/processed/accidents_clean.csv"

copy_df.to_csv(path, index=False, header=True)

print("The new cleaned dataset has been saved at dataset/processed/")
```
